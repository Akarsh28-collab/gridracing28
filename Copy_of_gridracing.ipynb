```python
"""
--------------------------------------------------------------------
GRID RACING AI ‚Äì Competing Pathfinding Agents (Short Version)
--------------------------------------------------------------------
Input Format:
Give grid rows & columns, then rows with:
A = Agent A start
B = Agent B start
G = Goal
# = Wall
. = Free cell

Output:
Time taken by both agents to reach the goal & final winner.
--------------------------------------------------------------------
"""

import heapq

# ------------- USER INPUT -------------
r, c = map(int, input("Enter rows and columns: ").split())
print("Enter grid row by row:")
grid = [list(input()) for _ in range(r)]

for i in range(r):
    for j in range(c):
        if grid[i][j] == "A": A = (i, j)
        if grid[i][j] == "B": B = (i, j)
        if grid[i][j] == "G": G = (i, j)

# ------------- A* PATHFINDING FUNCTION -------------
def astar(start):
    pq = [(0, start, 0)]    # (priority cost, node, travel_time)
    vis = set()
    while pq:
        _, (x, y), t = heapq.heappop(pq)
        if (x, y) == G: return t
        if (x, y) in vis: continue
        vis.add((x, y))
        for dx, dy in [(1,0),(-1,0),(0,1),(0,-1)]:
            nx, ny = x+dx, y+dy
            if 0 <= nx < r and 0 <= ny < c and grid[nx][ny] != "#":
                h = abs(G[0]-nx) + abs(G[1]-ny)  # heuristic: Manhattan distance
                heapq.heappush(pq, (t+1+h, (nx,ny), t+1))
    return float("inf")  # no path exists

# ------------- GAME SIMULATION -------------
timeA = astar(A)
timeB = astar(B)

print("\n----- GRID RACE RESULT -----")
print("Agent A time:", timeA)
print("Agent B time:", timeB)

if timeA < timeB: print("üèÜ WINNER ‚Üí Agent A")
elif timeB < timeA: print("üèÜ WINNER ‚Üí Agent B")
else: print("ü§ù DRAW ‚Äì same completion time")
```
